```python
import pandas as pd
import json
import random

# Function to read JSON file in chunks
def read_json_in_chunks(file_path, chunk_size=1000):
    with open(file_path, 'r') as file:
        while True:
            chunk = []
            for _ in range(chunk_size):
                line = file.readline()
                if not line:
                    return
                try:
                    chunk.append(json.loads(line))
                except json.JSONDecodeError:
                    continue
            yield chunk

# Function to take a random sample from each chunk
def take_random_sample_from_chunks(file_path, sample_size):
    sampled_data = []
    for chunk in read_json_in_chunks(file_path):
        sampled_data.extend(random.sample(chunk, min(sample_size, len(chunk))))
    return sampled_data

# Main function
def main():
    # Path to your JSON file
    json_file_path = 'Books_5.json'
    
    # Specify sample size
    sample_size = 100  # Change this to the desired sample size
    
    # Take a random sample from the JSON file
    sampled_data = take_random_sample_from_chunks(json_file_path, sample_size)
    
    # Optionally, write sampled data to a new JSON file
    with open('random_sample.json', 'w') as file:
        json.dump(sampled_data, file, indent=4)

if __name__ == "__main__":
    main()
```
